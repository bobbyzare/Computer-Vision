```python
img = cv2.imread("C:/Users/bobby/Downloads/Computer-Vision-with-Python/DATA/crossword.jpg",0)
```

```python
plt.imshow(img,cmap="gray")
```

```python
def show_pic(img,):
   fig = plt.figure(figsize=(15,15))
   ax = fig.add_subplot(111) 
   ax.imshow(img,cmap = "gray")
```

```python
show_pic(img)
```

```python
ret,th1 =cv2.threshold(img,200,255,cv2.THRESH_BINARY)
```

```python
show_pic(th1)
```

```python
th2 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,8)
```

```python
show_pic(th2)
```

```python
th3  = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,11,8)
```

```python
show_pic(th3)
```
