```python
face_cascade = cv2.CascadeClassifier("C:/Users/bobby/Downloads/Computer-Vision-with-Python/DATA/haarcascades/haarcascade_frontalface_default.xml")
```

```python
def detect_face(img):
     
        face_img = img.copy()
        
        face_rects = face_cascade.detectMultiScale(face_img)
        
        for (x,y,w,h) in face_rects:
            cv2.rectangle(face_img,(x,y),(x+w,y+h),(255,255,255),10)
            
        return face_img
```

```python
result = detect_face(denis)
```

```python
plt.imshow(result, cmap = "gray")
```

```python
result = detect_face(nadia)
```

```python
plt.imshow(result, cmap = "gray")
```

```python
result = detect_face(solvay)
```

```python
plt.imshow(result, cmap = "gray")
```

```python
def adj_detect_face(img):
     
        face_img = img.copy()
        
        face_rects = face_cascade.detectMultiScale(face_img,scaleFactor=1.2,minNeighbors=5)
        
        for (x,y,w,h) in face_rects:
            cv2.rectangle(face_img,(x,y),(x+w,y+h),(255,255,255),10)
            
        return face_img
```

```python
result = adj_detect_face(solvay)
```

```python
plt.imshow(result, cmap = "gray")
```

```python
eye_cascade = cv2.CascadeClassifier("C:/Users/bobby/Downloads/Computer-Vision-with-Python/DATA/haarcascades/haarcascade_eye.xml")
```

```python
def detect_eyes(img):
    
    face_img = img.copy()
  
    eyes = eye_cascade.detectMultiScale(face_img) 
    
    
    for (x,y,w,h) in eyes: 
        cv2.rectangle(face_img, (x,y), (x+w,y+h), (255,255,255), 5) 
        
    return face_img
    
```

```python
result = detect_eyes(nadia)
plt.imshow(result, cmap = "gray")
```

```python
result = detect_eyes(denis)
plt.imshow(result, cmap = "gray")
```

```python
result = detect_eyes(solvay)
plt.imshow(result, cmap = "gray")
```

```python
## face detection on live video
cap = cv2.VideoCapture(0) 

while True:
    
    ret,frame = cap.read(0)
    
    frame = detect_face(frame)
    
    cv2.imshow("video Face Detect",frame)
    
    k = cv2.waitKey(1)
    
    if k == 27:
        break
        
        
cap.release()
cv2.destroyAllWindows
```

```python
## eyes detection on live video

cap = cv2.VideoCapture(0)

while True:
    
    ret,frame = cap.read(0)
    
    frame = detect_eyes(frame)
    
    cv2.imshow("Eye Detection",frame)
    
    k = cv2.waitKey(1)
    
    if k == 27:
        break
        
cap.release()
cv2.destroyAllWindows
```
