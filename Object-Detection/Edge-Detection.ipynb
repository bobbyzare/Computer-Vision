```python
## EDGE DETECTION

import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

img = cv2.imread("C:/Users/bobby/Downloads/Computer-Vision-with-Python/DATA/sammy_face.jpg")
 #img = cv2.cvtColor(face,cv2.COLOR_BGR2RGB)

plt.imshow(img)

edges = cv2.Canny(image= img,threshold1=127,threshold2=127)

plt.imshow(edges)

edges = cv2.Canny(image= img,threshold1=220,threshold2=220)
plt.imshow(edges)

med_val = np.median(img)

med_val

# lowe threshold to either 0 or 70% of the median value whichever is greater
lower = int(max(0,0.7*med_val))
# upper threashold to either 130% of the median or the max 255, whichever is smaller
upper = int(min(0,0.7*med_val))

edges = cv2.Canny(image= img,threshold1=lower,threshold2=upper+100)
plt.imshow(edges)

blurred = cv2.blur(img,ksize=(5,5))

edges = cv2.Canny(image= blurred,threshold1=lower,threshold2=upper+50)
plt.imshow(edges)
```
